```python
from pathlib import Path

import matplotlib.pyplot as plt
import pandas as pd


PREFIX = ''
REPORT_DIR = Path.cwd() / 'reports' / PREFIX
```

#### Prefill stage

```python
df = pd.read_csv(
    REPORT_DIR / 'prefill-stage.csv',
    index_col=0,
)
```

```python
throughput = df['completion_tokens'] * df.index / df['elapsed_time']
plt.scatter(
    df.index,
    y=throughput,
    color='red', s=10,
)

plt.title('Prefill Throughput')
plt.xlabel('Number of requests')
plt.ylabel('tokens/sec')

lb, up = 0, 100*(throughput.max() // 100 + 1)
plt.ylim([lb, up])
plt.grid(color='grey', linestyle=':')
```

```python
plt.scatter(
    df.index,
    y=df['completion_tokens'] / df['elapsed_time'],
    color='red', s=10,
)

plt.title('Prefill Throughput (per request)')
plt.xlabel('Number of requests')
plt.ylabel('Prefill throughput, tokens/sec')

plt.grid(color='grey', linestyle=':')
```

```python
plt.scatter(
    df.index,
    y=df['elapsed_time'],
    color='red', s=10,
)


plt.title('Latency')
plt.xlabel('Number of requests')
plt.ylabel('TTFT, sec')

plt.ylim([0, None])
plt.grid(color='grey', linestyle=':')
```
