```python
from pathlib import Path

import matplotlib.pyplot as plt
import pandas as pd


PREFIX = ''
REPORT_DIR = Path.cwd().parent / 'reports' / PREFIX
```

#### Prefiling stage

```python
df = pd.read_csv(
    REPORT_DIR / 'prefilling-stage.csv',
    index_col=0,
)
```

```python
plt.scatter(
    df.index,
    y=df['completion_tokens'] * df.index / df['elapsed_time'],
    color='red', s=10,
)

plt.xlabel('Number of workers')
plt.ylabel('TPS, tokens/sec')
plt.title('Throughput (total)')

plt.grid(color='grey', linestyle=':')
```

```python
plt.scatter(
    df.index,
    y=df['completion_tokens'] / df['elapsed_time'],
    color='red', s=10,
)

plt.xlabel('Number of workers')
plt.ylabel('TPS, tokens/sec')
plt.title('Throughput (per worker)')

plt.grid(color='grey', linestyle=':')
```

```python
plt.scatter(
    df.index,
    y=df['elapsed_time'],
    color='red', s=10,
)

plt.ylim([0, None])

plt.xlabel('Number of workers')
plt.ylabel('TTFT, sec')
plt.title('Latency')

plt.grid(color='grey', linestyle=':')
```
